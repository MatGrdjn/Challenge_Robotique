```python
import pandas as pd
```

```python
results = pd.read_csv(r"C:\Users\Utilisateur\Documents\CoursCI2\Challenge\python_scripts\results\benchmark_results.csv")
```

```python
results.head()
```

```python



results["Gain_Reel"] = pd.to_numeric(results["Gain_Reel"])
results["Fuel_Reel"] = pd.to_numeric(results["Fuel_Reel"])

sorted_results = results.sort_values(by=["Map", "Gain_Reel", "Fuel_Reel"], ascending=[True, False, False])

meilleurs_algos = sorted_results.drop_duplicates(subset=["Map"], keep="first")

print(meilleurs_algos[["Map", "Algorithme", "Gain_Reel", "Fuel_Reel", "Chemin_Script"]])
```

```python
import os
import shutil

dossier_rendu = r"C:\Users\Utilisateur\Documents\CoursCI2\Challenge\python_scripts\rendu"
os.makedirs(dossier_rendu, exist_ok=True)

for index, row in meilleurs_algos.iterrows():
    nom_map = row["Map"]
    gain = int(row["Gain_Reel"])
    fuel = float(row["Fuel_Reel"])
    chemin_source = row["Chemin_Script"]
    
    if not os.path.isabs(chemin_source) and chemin_source.startswith("results"):
        chemin_source = os.path.join("..", chemin_source)
        
    nouveau_nom = f"{nom_map}_score_{gain}_fuel_{fuel:.1f}.txt"
    chemin_destination = os.path.join(dossier_rendu, nouveau_nom)
    
    try:
        shutil.copyfile(chemin_source, chemin_destination)
    except FileNotFoundError:
        print("FileNotFoundError")
```
